```python
f, ax = plt.subplots(figsize=(6, 15))

pays = data['Country Name']
# Plot 
sns.set_color_codes("pastel")
sns.barplot(x="Net enrolment rate, secondary, both sexes (%)", 
            y= pays, data=data,
            label="Net enrolment rate, secondary, both sexes (%)", color="b")

# Plot 
sns.set_color_codes("muted")
sns.barplot(x="Net enrolment rate, upper secondary, both sexes (%)", 
            y=pays, data=data,
            label="Net enrolment rate, upper secondary, both sexes (%)", color="b")
# Add a legend and informative axis label
ax.legend(ncol=2, loc=0, frameon=True)
ax.set(xlim=(0, 100), ylabel="",xlabel="Net enrolment rate, secondary and upper secondary, both sexes (%)")
sns.despine(right=True, top=True)
```
